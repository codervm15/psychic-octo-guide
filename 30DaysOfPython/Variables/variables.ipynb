Datatypes and Operators

```python
# assignment operator

a=2
b=4

c = 5

c += 4

print(c)

# arithmetic operator

print(a+b) #addition
print(a-b) #subtraction
print(a*b) #multiplication
print(a/b) #division
print(a%b) #modulus
print(a**b) #exponential
print(a//b) #floor div


```

```python
#logical operator
a = True
b = True

if(a and b):
    print(True)

x=True
y=False

if(x or y):
    print(True)
```

```python
#comparison operator

a=10
b=20

print(a==b)

if(a>b):
    print(True)
else:
    print(False)
```
