Revising List datatype , operators and methods

```python
#empty list

lst = []
```

```python
lst = ['banana', 'lemon', 'orange', 'apple', 'pear']

#list indexing
print(lst[0])

print(lst[0:4])

print(lst[-5:-2])

#list reverse
print(lst[::-1])
```

```python
lst1 = [1,2,3,4,5]

lst2 = [9,6,7,8]
lst1.extend(lst2)

print(lst1)

lst1.append("Anshul")

print(lst1)

print(len(lst1))

print(lst1.count(1))

print(lst2.pop())

lst2.sort()
print(lst2)

print(lst2.clear())

```

```python
#looping on list

fruits = ['banana', 'apple', 'mango']

for i in range(0, len(fruits)):
    print(fruits[i])


for x in fruits:
    print(x)
```
