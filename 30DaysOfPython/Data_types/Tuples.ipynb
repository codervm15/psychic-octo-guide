Revising Tuples datatype, operators, methods

```python
#empty tuple

t1 = ()

t2 = ('banana', 'mango','grapes')

print(t2)
```

```python
t3 = ('item1', 'item2', 'item3')

print(len(t3))

#tuple indexing

print(t3[0:2])
print(t3[-1:])

print(t3[::-1])
```

```python
#chaging tuple to list

fruits = ('banana', 'orange', 'mango', 'lemon')

fruits_lst = list(fruits)

print(fruits_lst)
```

```python
nordic_countries = ('Denmark', 'Finland','Iceland', 'Norway', 'Sweden')

print("india" in nordic_countries)
```
