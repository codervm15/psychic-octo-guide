### List comprehensions - fastest or shortest way to create list
basic syntax - [expression for iten in iterable]

with some logic - [expression for item in iterable if condition]

nested list comprehensions- [expression for item1 in iterable1 for item2 in iterable2]

```python
#create list of numnbers from 1 to 10

numbers = [number for number in range(1,11)]
print(numbers)
```

```python
#create list of squares of numbers from 1-10

squares = [number**2 for number in range(1,11)]
print(squares)
```

```python
#create a list of only even numbers from 1-20

even = [number for number in range(1,21) if number%2==0]
print(even)
```

```python
#nested list_comp

lst1 = [1,2,3,4,5,6,7]
lst2 = ['a','b','c','d']

lst3 = [(i,j) for i in lst1 for j in lst2] #it will create a list of tuple pairs
lst4 = [[i,j] for i in lst1 for j in lst2] #it will create a list of list pairs

print(lst3)
print(lst4)
```

```python
# Flattening a two dimensional array
list_of_lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

numbers = [number for row in list_of_lists for number in row]
print(numbers)
```

```python
#calculate length of each word in list

words = ['python', 'anshul', 'ram', 'gpu', 'asus', 'iphone']

length = [len(word) for word in words]
print(length)
```
