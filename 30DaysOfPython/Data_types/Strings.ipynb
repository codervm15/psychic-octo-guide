Revising Strings and it's operators, methods

```python
#By default everything in python is taken as string

name = 'Vikrant'

multiline = '''This is revising Python Playlist.
This is created by Vikrant'''

print(name)

print(multiline)
```

```python
#Escape characters

print("I hope you are enjoying the course. \nAre you?")

print("Days\tExercise\t\tTime")
print("1\tChest\t\t90mins")
print("2\tBack\t\t90mins")
print("3\tArms\t\t60mins")
print("4\tShoulder\t\t90mins")
print("5\tLegs\t\t90mins")
```

```python
#String formatting

first_name = "Vikrant"
last_name = "Mehla"
company = "AW"


print("My name is {} {} and I work for {}".format(first_name,last_name,company))

print(f'\nMy name is {first_name} {last_name} and I wokr for {company}')
```

```python
#string indexing

sent = "I love Python"

print(sent[0:5])

print(sent[-3:])

print(sent[-6:-1])

#string reversal
print(sent[::-1])
```

```python
char = "thirty days of python"

print(char.capitalize())

print(char.isalpha())

print(char.title())

print(char.upper())

print(char.lower())

print(char.endswith('on'))

print(char.find('ty'))

print(char.isalnum())

print(char.index('ys'))
```

```python
#Looping on strings

name = "Vikrant"

for i in range(0, len(name)):
    print(name[i])


for x in name:
    print(x)
```
