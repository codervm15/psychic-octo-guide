Simple Calculator

```python
n1 = int(input("Enter a number1"))
n2 = int(input("Enter a number2"))

operation = input("Enter an operation, such as +,-,*.**,/,%,//")

result = 0

print(n1)
print(n2)
print(operation)

if(operation == "+"):
    result = n1+n2
    print(result)
elif(operation=="-"):
    result= n1-n2
    print(result)
elif(operation=="*"):
    result= n1*n2
    print(result)
elif(operation=="/"):
    if(n2==0):
        print("Operation not possible due to divide by 0")
    else:
        result = n1/n2
        print(result)
elif(operation=="%"):
    result = n1%n2
    print(result)
elif(operation=="**"):
    result = n1**n2
    print(result)
elif(operation=="//"):
    result = n1//n2
    print(result)
else:
    print("Invalid Operation")
```
